##### 1. What is the total number of records in the dataset?


```python
print(" Total number of records in this dataset:", len(df))
```

##### 2. How many unique fetal health categories are present?

```python
unique_FH= df['fetal_health'].value_counts()
print("The unique fetal health categories in this dataset:", unique_FH)
```

##### 3. What is the standard deviation of the baseline value?

```python
print(df['baseline value'].std())
```

##### 4. What is the average baseline heart rate?


```python
average_baseline = df['baseline value'].mean()
print("The average baseline is:", average_baseline)
```

##### 5. What is the average histogram mean?

```python
average_histomean = df['histogram_mean'].mean()
print("The average histogram mean:", average_histomean)
```

##### 6.  What is the median histogram mode?



```python
histogram_mode_median = df['histogram_mode'].median()
print("Median histogram mode:", histogram_mode_median)
```

##### 7. What is the mean uterine contraction rate?


```python
uterine_contractions_mean = df['uterine_contractions'].mean()
print("Mean uterine contraction rate:", uterine_contractions_mean)
```

##### 8. What is the percentage of records with severe decelerations?


```python
severe_decelerations_count = len(df[df['severe_decelerations'] > 0])
severe_decelerations_percentage = (severe_decelerations_count / len(df)) * 100
print("Percentage of records with severe decelerations:", severe_decelerations_percentage)
```

##### 9. What is the correlation between the baseline value and fetal movement?

```python
cor = df['baseline value'].corr(df['fetal_movement'])
print("The correlation between baseline value and fetal movement", cor)
```

##### 10. What is the correlation between the baseline value and severe decelerations?

```python
print(df['baseline value'].corr(df['severe_decelerations']))
```

##### 11.  Compare the mean of all numerical columns across fetal health categories

```python
means_by_category = df.groupby('fetal_health').mean().sum()
print("Mean of all numerical columns by fetal health category:",means_by_category)
```

##### 12. Calculate the skewness of uterine contractions

```python
skewness_contractions = df['uterine_contractions'].skew()
print("Skewness of uterine contractions:", skewness_contractions)
```

##### 13. What is the relationship between the baseline value and histogram mode?

```python
plt.scatter(df['baseline value'], df['histogram_mode'])
plt.figure(figsize=(6,4))
plt.show()
```

##### 14. What is the distribution of prolongued decelerations?

```python
df['prolongued_decelerations'].value_counts().plot(kind='bar')
plt.show()
```

##### 15. Plot the distribution of baseline heart rate

```python
plt.figure(figsize=(4,2))
sns.histplot(df['baseline value'], kde=True, bins=20)
plt.title('Distribution of Baseline Heart Rate')
plt.xlabel('Baseline Value')
plt.ylabel('Frequency')
plt.show()
```

##### 16. Plot the count of each fetal health category.

```python

plt.figure(figsize=(4, 2))
sns.countplot(x='fetal_health', data=df, palette='viridis')
plt.title('Count of Fetal Health Categories')
plt.xlabel('Fetal Health')
plt.ylabel('Count')
plt.show()
```

##### 17. Plot a scatter plot of histogram variance vs histogram mode

```python

plt.figure(figsize=(4, 2))
sns.scatterplot(x='histogram_mode', y='histogram_variance', hue='fetal_health', data=df, palette='cool')
plt.title('Scatter Plot of Histogram Mode vs Variance')
plt.xlabel('Histogram Mode')
plt.ylabel('Histogram Variance')
plt.show()
```
